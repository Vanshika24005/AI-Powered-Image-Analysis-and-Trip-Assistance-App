# **Data Augmentation For CNN-based Model Prediction**

1. Import the neccessary libraries.

```python
!pip install icrawler
from icrawler.builtin import BingImageCrawler # Changed from GoogleImageCrawler
import os
import shutil
import random
from pathlib import Path
import time # Import time module
```

2. Demarcate Landmark Classes.

```python
classes = [
    # Asia
    "Taj Mahal",
    "Great Wall of China",
    "Angkor Wat",
    "Burj Khalifa",
    "Fushimi Inari Shrine",
    "Petra Jordan",
    "Bagan Temples Myanmar",

    # Europe
    "Eiffel Tower",
    "Colosseum Rome",
    "Big Ben London",
    "Sagrada Familia Barcelona",
    "Acropolis Athens",
    "Neuschwanstein Castle",

    # Americas
    "Statue of Liberty",
    "Machu Picchu",
    "Chichen Itza",
    "Christ the Redeemer",

    # Africa & Oceania
    "Pyramids of Giza",
    "Sydney Opera House",
    "Table Mountain Cape Town",
]
```

3. Configure and Normalize Images.

```python
IMAGES_PER_CLASS = 50
RAW_DIR         = Path("dataset/raw")
TRAIN_DIR       = Path("dataset/train")
VAL_DIR         = Path("dataset/val")
TRAIN_SPLIT     = 0.8   # 80% train, 20% val
```

4. Crawl the bing space for images.

```python
def crawl_images():
    for landmark in classes:
        save_dir = RAW_DIR / landmark
        save_dir.mkdir(parents=True, exist_ok=True)

        print(f"📸 Crawling: {landmark}")
        # Changed to BingImageCrawler
        crawler = BingImageCrawler(
            feeder_threads=2,
            parser_threads=2,
            downloader_threads=4,
            storage={"root_dir": str(save_dir)}
        )
        # Add a User-Agent to prevent being blocked (updated to a more recent Chrome version)
        crawler.downloader.session.headers.update({"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36"})
        try:
            crawler.crawl(
                keyword=f"{landmark} landmark",
                max_num=IMAGES_PER_CLASS,
                file_idx_offset=0,
            )
        except Exception as e:
            print(f"  ⚠️  Error crawling {landmark}: {e}")
        time.sleep(random.uniform(2, 5)) # Add a random delay between 2 and 5 seconds
```

5. Create Train/Val Split

```python
def split_dataset():
    print("\n📂 Splitting into train/val...")

    for landmark in classes:
        src_dir = RAW_DIR / landmark
        images  = list(src_dir.glob("*.jpg")) + \
                  list(src_dir.glob("*.png")) + \
                  list(src_dir.glob("*.jpeg"))

        if not images:
            print(f"  ⚠️  No images found for: {landmark}")
            continue

        random.shuffle(images)
        split_idx    = int(len(images) * TRAIN_SPLIT)
        train_images = images[:split_idx]
        val_images   = images[split_idx:]

        for subset, img_list in [("train", train_images), ("val", val_images)]:
            dest = (TRAIN_DIR if subset == "train" else VAL_DIR) / landmark
            dest.mkdir(parents=True, exist_ok=True)
            for img in img_list:
                shutil.copy(img, dest / img.name)

        print(f"  ✅ {landmark}: {len(train_images)} train | {len(val_images)} val")
```

6. Data Summarization.

```python
def summarize():
    print("\n📊 Dataset Summary:")
    print(f"  Classes : {len(classes)}")
    for split, base in [("Train", TRAIN_DIR), ("Val", VAL_DIR)]:
        total = sum(len(list((base / c).glob("*"))) for c in classes if (base / c).exists())
        print(f"  {split:5s}   : {total} images")
```

7. Implement Driver's Code.

```python
if __name__ == "__main__":
    crawl_images()
    split_dataset()
    summarize()
```
